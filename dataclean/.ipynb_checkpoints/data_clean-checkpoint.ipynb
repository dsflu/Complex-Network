```python
trip = pd.read_csv('trip.csv', encoding='utf8')
trip.head()
```

```python
max(trip["start_station_id"])
```

```python
import numpy as np
x=np.zeros((83, 83))
for i,j in zip(trip['start_station_id'],trip['end_station_id']):
    x[i-2,j-2]=x[i-2,j-2]+1;
```

```python
x_copy = x.copy();
range(1,len(x_copy))
```

```python
for i in range(len(x_copy)):
    for j in range(i+1,len(x_copy)):
        x_copy[i,j] = x_copy[i,j]+x_copy[j,i];
x_copy.max()
```

```python
jdata = [];

for i in range(len(x_copy)):
    for j in range(i+1,len(x_copy)):
        if x_copy[i,j] != 0:
            data = {};
            data['source'] = i;
            data['target'] = j;
            data['weight'] = x_copy[i,j];
            jdata.append(data);
    
jdata
```

```python
with open('trip_json.json', 'w') as f:
    json.dump(jdata, f)
```
